```python
import numpy as np
import pandas as pd
from pathlib import Path

DATA_DIR_CANDIDATES = [
    Path(r"D:\thao\d\datathon-2026-F.E.N.S\datathon-2026-round-1"),
]

for candidate in DATA_DIR_CANDIDATES:
    if candidate.exists():
        DATA_DIR = candidate
        break
else:
    raise FileNotFoundError(
        "Khong tim thay thu muc dataset. Hay cap nhat DATA_DIR_CANDIDATES cho dung may cua ban."
    )

print(f"Using dataset from: {DATA_DIR}")
```

```python
TABLE_FILES = {
    "products": "products.csv",
    "customers": "customers.csv",
    "promotions": "promotions.csv",
    "geography": "geography.csv",
    "orders": "orders.csv",
    "order_items": "order_items.csv",
    "payments": "payments.csv",
    "shipments": "shipments.csv",
    "returns": "returns.csv",
    "reviews": "reviews.csv",
    "sales": "sales.csv",
    "sample_submission": "sample_submission.csv",
    "inventory": "inventory.csv",
    "web_traffic": "web_traffic.csv",
}

DATE_COLUMNS = {
    "customers": ["signup_date"],
    "promotions": ["start_date", "end_date"],
    "orders": ["order_date"],
    "shipments": ["ship_date", "delivery_date"],
    "returns": ["return_date"],
    "reviews": ["review_date"],
    "sales": ["Date"],
    "inventory": ["snapshot_date"],
    "web_traffic": ["date"],
}


def load_tables(data_dir, table_files, date_columns):
    tables = {}
    for table_name, file_name in table_files.items():
        tables[table_name] = pd.read_csv(
            data_dir / file_name,
            parse_dates=date_columns.get(table_name),
        )
    return tables


def choose_option(result, options, mode="exact"):
    if mode == "exact":
        for answer_key, answer_value in options.items():
            if answer_value == result:
                return answer_key
        raise ValueError(f"Khong tim thay dap an exact cho {result!r}")

    if mode == "closest":
        option_series = pd.Series(options, name="option_value")
        return option_series.sub(result).abs().sort_values().index[0]

    raise ValueError(f"Unsupported mode: {mode}")


def record_answer(question, computed_result, answer_key, answer_value):
    answer_book[question] = {
        "question": question,
        "computed_result": computed_result,
        "selected_answer": f"{answer_key} ({answer_value})",
    }


def print_answer(question, metric_label, metric_value, answer_key, answer_value):
    print(f"{question} {metric_label}: {metric_value}")
    print(f"Answer: {answer_key} ({answer_value})")


tables = load_tables(DATA_DIR, TABLE_FILES, DATE_COLUMNS)
answer_book = {}

print("=== DATASET OVERVIEW ===")
for table_name, table_df in tables.items():
    print(f"{table_name:20s} shape = {table_df.shape}")
```

### Q1. Trong số các khách hàng có nhiều hơn một đơn hàng, trung vị số ngày giữa hai lần mua liên tiếp (inter-order gap) xấp xỉ là bao nhiêu? (Tính từ orders.csv)

- Sắp xếp `orders` theo `customer_id`, `order_date`, `order_id`
- Với mỗi khách hàng, lấy ngày của đơn trước bằng `shift()`
- Tính số ngày chênh lệch giữa hai đơn liên tiếp
- Lấy trung vị của toàn bộ các khoảng cách này


```python
# Q1
q1_options = {"A": 30, "B": 90, "C": 144, "D": 365}

q1_orders = tables["orders"].sort_values(["customer_id", "order_date", "order_id"]).copy()
q1_orders["prev_order_date"] = q1_orders.groupby("customer_id")["order_date"].shift()
q1_orders["inter_order_gap_days"] = (
    q1_orders["order_date"] - q1_orders["prev_order_date"]
).dt.days

q1_gap_days = q1_orders["inter_order_gap_days"].dropna()
q1_result = q1_gap_days.median()
q1_answer = choose_option(q1_result, q1_options, mode="closest")

q1_preview = q1_orders.loc[
    q1_orders["inter_order_gap_days"].notna(),
    ["customer_id", "order_date", "prev_order_date", "inter_order_gap_days"],
].head(10)
q1_preview
```

```python
record_answer("Q1", f"{q1_result:.0f} days", q1_answer, f"{q1_options[q1_answer]} days")
print_answer("Q1", "median inter-order gap", f"{q1_result:.0f} days", q1_answer, f"{q1_options[q1_answer]} days")

q1_gap_days.describe(percentiles=[0.25, 0.5, 0.75]).to_frame("inter_order_gap_days")
```

### Q2. Phân khúc sản phẩm (segment) nào trong products.csv có tỷ suất lợi nhuận gộp trung bình cao nhất, với công thức (price − cogs)/price?

- Với từng sản phẩm, tính biên lợi nhuận gộp: `(price - cogs) / price`
- Gom nhóm theo `segment`
- Lấy giá trị trung bình của từng nhóm
- Segment nào có trung bình cao nhất thì chọn


```python
# Q2
q2_options = {"A": "Premium", "B": "Performance", "C": "Activewear", "D": "Standard"}

q2_products = tables["products"].copy()
q2_products["gross_profit_margin"] = (
    q2_products["price"] - q2_products["cogs"]
) / q2_products["price"]

q2_result = (
    q2_products.groupby("segment")["gross_profit_margin"]
    .mean()
    .sort_values(ascending=False)
)
q2_top_segment = q2_result.index[0]
q2_answer = choose_option(q2_top_segment, q2_options, mode="exact")

q2_result.to_frame("avg_gross_profit_margin")
```

```python
record_answer("Q2", q2_top_segment, q2_answer, q2_options[q2_answer])
print_answer("Q2", "top segment", q2_top_segment, q2_answer, q2_options[q2_answer])
```

### Q6. Trong customers.csv, xét các khách hàng có age_group khác null, nhóm tuổi nào có số đơn hàng trung bình trên mỗi khách hàng cao nhất? (tổng số đơn / số khách hàng trong nhóm)

- Đếm tổng số đơn của từng `customer_id` từ `orders`
- Join sang `customers` để lấy `age_group`
- Với mỗi nhóm tuổi, tính `tong so don / tong so khach hang`
- Nhóm nào có giá trị cao nhất thì chọn


```python
# Q6
q6_options = {"A": "55+", "B": "25-34", "C": "35-44", "D": "45-54"}

q6_order_count = tables["orders"].groupby("customer_id").size().rename("order_count")
q6_customers = tables["customers"][["customer_id", "age_group"]].merge(
    q6_order_count,
    on="customer_id",
    how="left",
)
q6_customers["order_count"] = q6_customers["order_count"].fillna(0)

q6_result = (
    q6_customers.loc[q6_customers["age_group"].notna()]
    .groupby("age_group")
    .agg(total_orders=("order_count", "sum"), total_customers=("customer_id", "count"))
)
q6_result["avg_orders_per_customer"] = (
    q6_result["total_orders"] / q6_result["total_customers"]
)
q6_result = q6_result.sort_values("avg_orders_per_customer", ascending=False)
q6_top_group = q6_result.index[0]
q6_answer = choose_option(q6_top_group, q6_options, mode="exact")

record_answer("Q6", q6_top_group, q6_answer, q6_options[q6_answer])
print_answer("Q6", "highest average orders/customer", q6_top_group, q6_answer, q6_options[q6_answer])
q6_result
```

### Q7. Vùng (region) nào trong geography.csv tạo ra tổng doanh thu cao nhất trong sales_train.csv?

`sales.csv` chỉ cho doanh thu theo ngày, không có `region`, nên cần suy ngược từ dữ liệu giao dịch.

- Tính doanh thu từng dòng trong `order_items`: `quantity * unit_price - discount_amount`
- Cộng lên theo `order_id` để ra doanh thu mỗi đơn
- Join với `orders` để lấy `order_date` và `zip`
- Lọc đúng khoảng thời gian train của `sales.csv`
- Join với `geography` để lấy `region`
- Group theo `region` và cộng doanh thu


```python
# Q7
q7_options = {"A": "West", "B": "Central", "C": "East", "D": "Ca ba vung xap xi bang nhau"}

q7_train_start = tables["sales"]["Date"].min()
q7_train_end = tables["sales"]["Date"].max()

q7_order_revenue = (
    tables["order_items"]
    .assign(line_revenue=lambda df: df["quantity"] * df["unit_price"] - df["discount_amount"])
    .groupby("order_id", as_index=False)["line_revenue"]
    .sum()
)

q7_region_revenue = (
    q7_order_revenue
    .merge(tables["orders"][["order_id", "order_date", "zip"]], on="order_id", how="left")
    .loc[lambda df: df["order_date"].between(q7_train_start, q7_train_end)]
    .merge(tables["geography"][["zip", "region"]], on="zip", how="left")
    .groupby("region")["line_revenue"]
    .sum()
    .sort_values(ascending=False)
)
q7_top_region = q7_region_revenue.index[0]
q7_answer = choose_option(q7_top_region, q7_options, mode="exact")

record_answer("Q7", q7_top_region, q7_answer, q7_options[q7_answer])
print(f"Q7 train window: {q7_train_start.date()} -> {q7_train_end.date()}")
print_answer("Q7", "highest revenue region", q7_top_region, q7_answer, q7_options[q7_answer])
q7_region_revenue
```

### Q9. Trong bốn kích thước sản phẩm (S, M, L, XL), kích thước nào có tỷ lệ trả hàng cao nhất, được định nghĩa là số bản ghi trong returns chia cho số dòng trong order_items (join với products theo product_id)?

Theo đề, return rate được tính như sau:

- Tử số: số bản ghi trong `returns` theo từng `size`
- Mẫu số: số dòng trong `order_items` theo từng `size`
- Hai bảng này đều phải join với `products` để lấy `size`
- Sau đó lấy `returns / order_items` và so sánh 4 size `S, M, L, XL`


```python
# Q9
q9_options = {"A": "S", "B": "M", "C": "L", "D": "XL"}

q9_size_map = tables["products"][["product_id", "size"]]
q9_returns_by_size = (
    tables["returns"][["product_id"]]
    .merge(q9_size_map, on="product_id", how="left")["size"]
    .value_counts()
    .rename("return_records")
)
q9_order_items_by_size = (
    tables["order_items"][["product_id"]]
    .merge(q9_size_map, on="product_id", how="left")["size"]
    .value_counts()
    .rename("order_item_rows")
)

q9_result = pd.concat([q9_returns_by_size, q9_order_items_by_size], axis=1).fillna(0)
q9_result["return_rate"] = q9_result["return_records"] / q9_result["order_item_rows"]
q9_result = q9_result.loc[["S", "M", "L", "XL"]].sort_values("return_rate", ascending=False)
q9_top_size = q9_result.index[0]
q9_answer = choose_option(q9_top_size, q9_options, mode="exact")

record_answer("Q9", q9_top_size, q9_answer, q9_options[q9_answer])
print_answer("Q9", "highest return-rate size", q9_top_size, q9_answer, q9_options[q9_answer])
q9_result
```

#### 5. Size `S` có tỷ lệ trả hàng cao nhất trong bốn size S/M/L/XL

Trong bốn size `S`, `M`, `L`, `XL`, size `S` có tỷ lệ trả hàng cao nhất theo định nghĩa của đề:

`số bản ghi returns / số dòng order_items`

Kết quả này củng cố thêm nhận định rằng sizing là một vấn đề cần được ưu tiên phân tích. Size `S` cần được kiểm tra kỹ hơn để xem nguyên nhân đến từ bảng size chưa chính xác, form sản phẩm không đồng nhất, hay kỳ vọng của khách hàng chưa khớp với sản phẩm thực tế.

**Ý nghĩa kinh doanh:**  
Doanh nghiệp nên ưu tiên kiểm tra lại size guide và fit của size `S`, đặc biệt ở các danh mục có tỷ lệ trả hàng cao.

---

#### 6. `email_campaign` mang lại traffic có chất lượng tốt hơn

Nguồn `email_campaign` có bounce rate trung bình thấp nhất trong các nguồn traffic.

Bounce rate thấp cho thấy người dùng đến từ email có khả năng quan tâm đến thương hiệu hoặc sản phẩm cao hơn so với một số kênh traffic đại trà. Đây có thể là nhóm khách đã biết đến thương hiệu, đã từng mua hoặc có mức độ quan tâm cao hơn.

**Ý nghĩa kinh doanh:**  
Email nên được xem là kênh có intent cao, phù hợp cho retention, reactivation, giới thiệu sản phẩm cá nhân hóa và các chiến dịch nhắc mua lại.

---

#### 7. Khuyến mãi là một phần quan trọng trong cơ chế bán hàng

Khoảng `39%` dòng trong `order_items` có áp dụng khuyến mãi, tức là `promo_id` không null.

Điều này cho thấy khuyến mãi không phải là trường hợp hiếm, mà là một phần đáng kể trong hoạt động bán hàng. Vì vậy, doanh nghiệp không nên chỉ nhìn doanh thu trong thời gian khuyến mãi, mà cần đánh giá thêm hiệu quả khuyến mãi đối với margin, tỷ lệ trả hàng và lợi nhuận thực tế.

**Ý nghĩa kinh doanh:**  
Doanh nghiệp nên phân tích `Promo ROI`, thay vì chỉ đo doanh thu do khuyến mãi tạo ra. Một chương trình khuyến mãi chỉ thật sự hiệu quả nếu giúp tăng doanh số có lợi nhuận, không làm biên lợi nhuận bị bào mòn quá nhiều.

---

#### 8. Vùng `East` tạo ra tổng doanh thu cao nhất

Vùng `East` tạo ra tổng doanh thu cao nhất trong tập train.

Điều này cho thấy `East` có thể là khu vực doanh thu cốt lõi của doanh nghiệp. Tuy nhiên, doanh thu cao chưa đủ để kết luận đây là khu vực nên đầu tư mạnh nhất. Cần phân tích thêm margin, tỷ lệ trả hàng, tỷ lệ hủy đơn và chi phí vận chuyển theo vùng.

**Ý nghĩa kinh doanh:**  
Vùng `East` là khu vực tự nhiên để ưu tiên trong các chiến dịch tăng trưởng, nhưng cần kiểm chứng thêm xem khu vực này có thật sự hiệu quả về lợi nhuận và vận hành hay không.

---

#### 9. Trong các đơn bị hủy, `credit_card` xuất hiện nhiều nhất

Trong nhóm đơn hàng có trạng thái `cancelled`, phương thức thanh toán xuất hiện nhiều nhất là `credit_card`.

Tuy nhiên, không nên kết luận ngay rằng `credit_card` gây ra hủy đơn nhiều hơn. Lý do là nếu `credit_card` vốn là phương thức thanh toán phổ biến nhất trong toàn bộ dữ liệu, thì việc nó xuất hiện nhiều nhất trong nhóm đơn bị hủy cũng là điều dễ hiểu.

Điều cần làm tiếp theo là tính tỷ lệ hủy theo từng phương thức thanh toán:

`cancel_rate = số đơn cancelled / tổng số đơn`

**Ý nghĩa kinh doanh:**  
Cần phân tích sâu hơn để xác định liệu `credit_card` có thật sự có rủi ro hủy cao hơn hay chỉ xuất hiện nhiều vì đây là phương thức thanh toán phổ biến.

---

#### 10. Kế hoạch trả góp `6 kỳ` có giá trị thanh toán trung bình cao nhất

Các đơn hàng thanh toán theo hình thức trả góp `6 kỳ` có giá trị thanh toán trung bình cao nhất.

Điều này gợi ý rằng các đơn hàng có giá trị lớn có xu hướng gắn với kế hoạch trả góp trung hạn. Khách hàng có thể muốn chia nhỏ khoản thanh toán khi giá trị đơn hàng cao, nhưng không nhất thiết cần kéo dài đến 12 kỳ.

**Ý nghĩa kinh doanh:**  
Doanh nghiệp có thể thử nghiệm các chiến dịch khuyến khích trả góp cho giỏ hàng giá trị cao, ví dụ “trả góp 0% cho đơn hàng trên một ngưỡng nhất định”, đặc biệt với các sản phẩm premium hoặc combo giá trị lớn.

### Các hướng phân tích chiến lược ban đầu

Từ kết quả 10 câu trắc nghiệm, có thể rút ra 3 hướng phân tích chính để phát triển tiếp trong Phần 2.

---

#### 1. Tối ưu trải nghiệm size/fit

Việc `wrong_size` là lý do trả hàng phổ biến nhất trong nhóm `Streetwear`, đồng thời size `S` có tỷ lệ trả hàng cao nhất, cho thấy sizing có thể là một vấn đề quan trọng trong trải nghiệm khách hàng.

**Các phân tích nên đào sâu:**

- Tỷ lệ trả hàng theo `category × size`
- Lý do trả hàng theo `category`
- Tỷ lệ trả hàng theo `product_id`
- Tỷ lệ trả hàng giữa đơn có promo và không có promo
- Rating trung bình theo size/category

**Hành động tiềm năng:**

- Cải thiện size guide
- Thêm gợi ý size
- Hiển thị chiều cao/cân nặng của người mẫu
- Cho phép lọc review theo size
- Tránh đẩy mạnh các SKU có tỷ lệ trả hàng cao trước khi xử lý vấn đề sizing

---

#### 2. Xây dựng chiến lược CRM quanh chu kỳ mua lại `144 ngày`

Trung vị inter-order gap khoảng `144 ngày` tạo ra một mốc thời gian rõ ràng để kích hoạt lại khách hàng.

**Các phân tích nên đào sâu:**

- Cohort retention theo tháng đăng ký
- RFM segmentation
- Repeat purchase rate theo age group
- Inter-order gap theo acquisition channel
- Khả năng win-back theo từng nhóm khách hàng

**Hành động tiềm năng:**

- Gửi reminder quanh ngày 120–150
- Kích hoạt win-back offer cho khách không mua sau 150 ngày
- Tạo loyalty benefit cho nhóm khách mua thường xuyên
- Cá nhân hóa email campaign dựa trên lịch sử mua hàng

---

#### 3. Ưu tiên các phân khúc, kênh và khu vực có giá trị cao

Kết quả ban đầu cho thấy nhóm `55+`, kênh `email_campaign` và vùng `East` là những điểm đáng chú ý.

**Các phân tích nên đào sâu:**

- AOV và repeat purchase rate theo age group
- Revenue và margin theo region
- Chất lượng chuyển đổi theo traffic source
- Tỷ lệ hủy đơn và trả hàng theo customer segment
- LTV proxy theo acquisition channel

**Hành động tiềm năng:**

- Dùng `email_campaign` cho retention và reactivation
- Ưu tiên vùng `East` nếu margin và return rate cũng tốt
- Thiết kế chương trình retention riêng cho nhóm `55+`
- Dùng ưu đãi trả góp cho các đơn hàng giá trị cao
